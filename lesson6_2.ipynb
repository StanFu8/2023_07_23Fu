```python
start=int(input("起始"))
end=int(input(" 終值"))

oddtotal=0
eventotal=0
for x in range(start,end+1):
    if x %2==0:
        eventotal+=x
    else:
        oddtotal+=x
print(f"從{start}到{end}\n奇總:{oddtotal:,}\n偶總:{eventotal}")
```

```python
for i in range(1,10):
    print(f"i={i}")
    for j in range(1,10):
        print(f"j={j}",end=" ")
    print()
```

```python
for i in range(1,10):    
    for j in range(1,10):
        print(f"{j:>d}*{i:<d}={j*i:>2d}",end=' ')
    print()
```

```python
for i in range(1,101):
    if i==1:
        continue
    print(i,end=" ")
print()
```

```python
for i in range(1,101):
    is_prime=True
    if i==1:
        continue
    for j in range(2,i):
        if i%j==0:
            is_prime=False
            break
    if is_prime:
        print(i,end=" ")

print()
```

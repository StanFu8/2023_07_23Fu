```python
#標準寫法
import matplotlib.pyplot as plt

# 1.製作figure
fig = plt.figure()

# 2.在1.的figure上製作一個axes
ax_1 = fig.add_subplot(1, 1, 1)

# 3.將圖的數據儲存在axes中
X = [0, 5, 6]
Y = [3, 4, 5]
ax_1.plot(X, Y)

# 4.顯示圖
plt.show()
```

```python
figure = plt.figure()
ax1 = figure.add_subplot(1,3,1)
ax1.set_title("Axes1")
ax2 = figure.add_subplot(1,3,2)
ax2.set_title("Axes2")
ax3 = figure.add_subplot(1,3,3)
ax3.set_title("Axes3")

plt.show()
```

```python
figure = plt.figure(figsize=(8,3),dpi=80)
stu1 = {'chinese':80, 'english':92, 'math':75,'history': 89, 'created':69}
stu2 = {'chinese':78, 'english':48, 'math':62,'history': 99, 'created':70}

axes = figure.add_subplot()
axes.plot(stu1.keys(),stu1.values())
axes.plot(stu2.keys(),stu2.values())
plt.show()
```
